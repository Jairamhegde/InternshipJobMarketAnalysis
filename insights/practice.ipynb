```python
import pandas as pd
df=pd.read_csv("cleaned.csv")
print(df['job'].value_counts().head())


df.columns
```

```python
salary_stats = (
    df.groupby("job")
      .agg(
          avg_salary=("Max_sal", "mean"),
          min_salary=("Min_sal", "min"),
          max_salary=("Max_sal", "max")
      )
      .reset_index()
      .sort_values(by="avg_salary", ascending=False)
)

print(salary_stats.head(10))
```

```python
import re
from datetime import datetime,timedelta

string = ["3 day ago","2 weeks ago","4 days ago","1 weeks ago"]

# for i in string :
def dateFromtext(i):
    match = re.search(r"(\d+)\s(\w+)\sago",i)
    value = int(match.group(1))
    measure = match.group(2)
    if measure == "day" or measure == "days":
        date = (datetime.now() - timedelta(days=value)).strftime("%Y-%m-%d")
        print(date)
    if measure == "week" or measure == "weeks":
        wdate = (datetime.now() - timedelta(weeks=value)).strftime("%Y-%m-%d")
        print(wdate)
# print(day)

# print(re.search(r"\d+",string).group())
```
